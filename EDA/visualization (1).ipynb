## Visualization

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x="Price", y="discount", data = df, hue="annual_electricity_Consumption")
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(x="Price",y="annual_electricity_Consumption",data=df,hue="Color")
plt.show()
```

```python
cross_tab = pd.crosstab(index=df['Brand'],columns=df['Color'], values=df['Price'],aggfunc="mean")
cross_tab.plot(kind='bar', stacked=True, colormap='viridis')
plt.title("Average Price by Brand and Color")
plt.xlabel("Brand")
plt.ylabel("Average Price")
plt.xticks(rotation=90) 
plt.show()
```

```python
plt.figure(figsize=(8, 6))  
sns.violinplot(x=df["Price"],bins=10, color="skyblue")  

plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Density")
plt.grid()
plt.show()
```

```python
sns.boxplot(x=df3,y=df.rating)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,5))

ax=df.groupby("Brand")["Price"].agg(["min","max"]).plot(kind="bar")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=5, color='black')
plt.tight_layout()
plt.show()
```
